```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread('Program-1.jpeg')
plt.imshow(image,cmap='gray')
plt.show()
```

```python
# Define three different kernels
mask_x = np.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

mask_y = np.array([[-1, -2, -1],
                     [ 0,  0,  0],
                     [ 1,  2,  1]])

mask_gaussian= (1/16) * np.array([[1, 2, 1],
                                   [2, 4, 2],
                                   [1, 2, 1]])
```

```python
# QUE1] Apply correlation  (no flipping of kernel)
corr_mask_x = cv2.filter2D(image, -1, mask_x)
corr_mask_y = cv2.filter2D(image, -2, mask_y)
corr_gaussian = cv2.filter2D(image, -2, mask_gaussian) 
```

```python
# QUE2] Flip kernels by 180 degrees for convolution
conv_mask_x = cv2.filter2D(image, -1, np.flip(mask_x))
conv_mask_y = cv2.filter2D(image, -1, np.flip(mask_y))
conv_gaussian = cv2.filter2D(image, -1, np.flip(mask_gaussian))
```

```python
# Displaying results
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
axes[0, 0].imshow(corr_mask_x, cmap='gray'); axes[0, 0].set_title('Correlation - Mask X')
axes[0, 1].imshow(corr_mask_y, cmap='gray'); axes[0, 1].set_title('Correlation - Mask Y')
axes[0, 2].imshow(corr_gaussian, cmap='gray'); axes[0, 2].set_title('Correlation - Gaussian')

axes[1, 0].imshow(conv_mask_x, cmap='gray'); axes[1, 0].set_title('Convolution - Mask X')
axes[1, 1].imshow(conv_mask_y, cmap='gray'); axes[1, 1].set_title('Convolution - Mask Y')
axes[1, 2].imshow(conv_gaussian, cmap='gray'); axes[1, 2].set_title('Convolution - Gaussian')

for ax in axes.ravel():
    ax.axis('off')
plt.tight_layout()
plt.show()
```
