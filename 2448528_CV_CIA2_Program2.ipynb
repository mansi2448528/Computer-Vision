```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import minimum_filter
from scipy.ndimage import maximum_filter
```

```python
image_n=cv2.imread('saltnoise.webp')
plt.imshow(image_n,cmap='gray')
plt.show() 
```

```python
image_p=cv2.imread('peppernoise.webp')
plt.imshow(image_p)
plt.show()
```

```python
image_np=cv2.imread('imagewithsalt and pepper noise.png')
plt.imshow(image_np)
plt.show() 
```

```python
# QUE1] SALT NOISE-MINIMUM FILTER 
# Apply minimum filter with different kernel sizes
kernel_sizes = [3, 5, 7]
plt.figure(figsize=(10, 20))
for i, k in enumerate(kernel_sizes):
    min_filtered = minimum_filter(image_n, size=k)
    plt.subplot(1, len(kernel_sizes), i+1)
    plt.imshow(min_filtered, cmap="gray")
    plt.title(f"Min Filter {k}x{k}")
    plt.axis("off")

plt.show() 
```

```python
# QUE2] PEPPER NOISE-MAXIMUM FILTER 
kernel_sizes = [3, 5, 7]
plt.figure(figsize=(10, 10))
for i, k in enumerate(kernel_sizes):
    max_filtered = maximum_filter(image_p, size=k)
    plt.subplot(1, len(kernel_sizes), i+1)
    plt.imshow(max_filtered, cmap="gray")
    plt.title(f"Max Filter {k}x{k}")
    plt.axis("off")
plt.show()
```

```python
# QUE3] SALT AND PEPPER NOISE - MEDIAN FILTER
def apply_median_filter(image_np, kernel_size):
    return cv2.medianBlur(image_np, kernel_size)
    
# Apply median filtering with different kernel sizes
median_3x3 = apply_median_filter(image_np, 3)
median_5x5 = apply_median_filter(image_np, 5)
median_7x7 = apply_median_filter(image_np, 7)
fig, axes = plt.subplots(4, 3, figsize=(12, 16)) 

axes[2, 0].imshow(median_3x3, cmap='gray'); axes[2, 0].set_title('Median Filter 3x3')
axes[2, 1].imshow(median_5x5, cmap='gray'); axes[2, 1].set_title('Median Filter 5x5')
axes[2, 2].imshow(median_7x7, cmap='gray'); axes[2, 2].set_title('Median Filter 7x7')


for ax in axes.ravel():
    ax.axis('off')
plt.tight_layout()
plt.show() 
```

```python
# QUE 4] MID POINT 
def midpoint_filter(image_np, ksize):
    min_filtered = minimum_filter(image_np, size=ksize)
    max_filtered = maximum_filter(image_np, size=ksize)
    return (min_filtered + max_filtered) // 2

# Apply midpoint filter
plt.figure(figsize=(10, 5))

for i, k in enumerate(kernel_sizes):
    midpoint_filtered = midpoint_filter(image_np, k)
    plt.subplot(1, len(kernel_sizes), i+1)
    plt.imshow(midpoint_filtered, cmap="gray")
    plt.title(f"Midpoint Filter {k}x{k}")
    plt.axis("off")

plt.show() 
```
